# Monte Carlo Simulation

approximating integrals by using the law of large numbers, that is sample mean is equal to the expected value(integral):

1.  generate random samples from the uniform dist. $X=$ $X_1$, $X_2$, $X_3$...
2. calculate $Y=g(X)=[g(X_1),g(X_2)...g(X_n)]$
3. now $\int_{a}^{b}g(x)dx = \frac{b-a}{N} \sum_{i=1}^{N}g(X_i)$

here R.H.S is $(b-a)$ times sample mean and L.H.S is the expected value of the R.V $g(X)$. when N is large this equality is nearly true.

```python
a=0
b=1
N=10000
```

```python
samples= np.random.uniform(a,b,N)  
g = lambda x:np.exp(-x)         # defining the function for which we want to evaluate the integral
value = ((b-a)/N)*(sum(g(samples)))   #(b-a) times sample mean equals value of definite integral between a and b 
value
```

# importance sampling

* while evaluating integral for a function, identify part of the function that looks like a well known pdf.
* say we have, $$\int_{a}^{b}\phi(x)dx$$ $$\implies\int_{a}^{b}f(x)g(x)dx= E[g(x)] $$
,where $f(x)$ is the known pdf.
* so now instead of generating random samples from uniform we generate from $f(x)$ distribution. this takes off compuatation load because generation of random samples more or less takes the same time. but due to this the seconfd step of evaluating function values for $g(x)$ is simplified

$$f(x; \frac{1}{\beta}) = \frac{1}{\beta} \exp(-\frac{x}{\beta})$$
          
for $x > 0$ and 0 elsewhere. :$\beta$ is the scale parameter,
which is the inverse of the rate parameter $\lambda = 1/\beta$.
The rate parameter is an alternative, widely used parameterization
of the exponential distribution [3]_.

```python
N = 100000
b=np.inf
a=0
samp = lambda size: np.random.exponential(1,size)  # sampling from f(x)
samp_trunc = lambda dist: [sam for sam in dist if a<=sam<=b]  
# tried limiting sampled values but doesn't work , probably because that
# is not a valid distribution, proper truncated distribution needed. 
# taking large values of b works tho, as expected
gun = lambda x: np.sin(x) #g(x)                                
val = lambda N: np.mean(gun(samp_trunc(samp(N)))) 
val(N)
```
