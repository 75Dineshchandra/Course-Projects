# Monte Carlo Simulation

approximating integrals by using the law of large numbers, that is sample mean is equal to the expected value(integral):

1.  generate random samples from the uniform dist. $X=$ $X_1$, $X_2$, $X_3$...
2. calculate $Y=g(X)=[g(X_1),g(X_2)...g(X_n)]$
3. now $\int_{a}^{b}g(x)dx = \frac{b-a}{N} \sum_{i=1}^{N}g(X_i)$

here R.H.S is $(b-a)$ times sample mean and L.H.S is the expected value of the R.V $g(X)$. when N is large this equality is nearly true.

```python
a=0
b=1
N=10000
```

```python
samples= np.random.uniform(a,b,N)  
g = lambda x:np.exp(-x)         # defining the function for which we want to evaluate the integral
value = ((b-a)/N)*(sum(g(samples)))   #(b-a) times sample mean equals value of definite integral between a and b 
value
```

 # contd. from optimization
 
 ## constrained optimization
 
 $min f(x)$, subject to $g(x)=c$ (constraint)
 1. use in economics
 2. communication, $\frac{1}{2}log(1+SNR)$. we want to maximize the rate but subject to some constrained SNR.
 3. linear regression, given $(X_1, X_2, X_2...X_i), (Y_1, Y_2, Y_3...Y_i)$ as the data set, we want to wint a line fitting this data as $ Y=aX + b$. the goal is to find the best fitting value of $a$ and $b$.
 
* we find $Y_{i_{predicted}}= aX_i + b$ for all value of the data set.

*  then calculate the Mean Square Error(Loss function) as, $$\frac{1}{N} \sum_{i=1}^N{(Y_i - Y_{i_{predicted}})}^2$$

* taking in more dimention to relate with the labels($y_i$) we can use more input features(X_i) at once.

* to prevent overfitting or underfitting we sometimes need to put constraints on the value of $a$ and $b$ and also to reach the optimal model have to minimize the loss function therefore we have a constrained optimization problem.


### back to general constrained optimization.

* $f(x,y) = 2x + y$, $g(x,y)= x^2 +y^2$
* when a function is made equal to a constant(parameter) we can think of it as a contour for a higher dimentional topology, like circle put equal to a constant(parameter) can be thought of a different sections of a sphere. or a line made equal to a parameter can be seen as section of a plane.

so we have, $$f(x,y)=2x+y$$,$$g(x,y)=x^2+y^2=c$$

we take $c=1$ for now. 

therefore, we have, $$min f(x,y)=2x+y$$, subject to $$g(x,y)=x^2+y^2=1$$

The point of minima or maxima will be the point where the contours of the touch, therefore at these point the slope of the two contours has to be i=the same that is the gradient of f and that of g have to be parallel to each other

$$\nabla f(x_0, y_0) = \lambda \nabla g(x_0, y_0)$$

where $\lambda$ is called the lagrange multiplier.

$$L = f(x,y) - \lambda[g(x,y) -c]$$

$L$ is called the lagrange function.


therefore the minimizing condition becomes 
, $$ \nabla L = 0$$

. solving for the current problem, we get three equations 

$$ 2x\lambda = 2$$
$$2y\lambda = 1$$
$$x^2 + y^2 = 1$$



solving these equation we get $$x= \frac{2}{\sqrt{5}}$$ $$y= \frac{1}{\sqrt{5}}$$, $$\lambda= \frac{\sqrt{5}}{2}$$ 



# 23 may

for more than one variables we use a vector for the variables.

$$X =[X_0, X_1, X_2....X_i]$$

and we want to find, 

$$ min f(X)$$

remember: for contour of a function we equate it to a constant(parameter).

1. we write the contour for $f(X)$ by equating it to some $k$.
2. we want the minimum value of k for which the contour touches the constraint curve.
3. at this point the tangent of the contour curve and that of the constraint curve will be parallel. 
4. hence the condition, $$\nabla f(X) = \lambda \nabla (g(X)-c)$$.
5. this converts to gradient of the lagrange function being zero.  $$ \nabla L = 0$$ where, $$L = f(X) - \lambda[g(X) -c]$$

therefore the algo is,

1. identify function to minimize(objective function).
2. identify the constraints( can be equality or inequality constarints)
3. check for bounds
4. choose "Method"( which optimizer to use, COBYLA? ADAM? etc)

## Inequality constraints

now in addition to the equality constraints and the bounds we also have inequality constraints on the variables.

so the problem reduces to , 
$$ min f(X)$$ subject to$$g(X)-c=0$$ and $$h(X)\geq 0$$( due to python condition of inequality constraints always taking the greater than equal to sign.

Incoporating equality constraintsin the Lagrange function using the $\lambda$ coeffecients is called **Regularization**. we can do the same thing for the inequality constraints and use a set of $\mu$ coefficients for each constraint.
$$L = f(X) + \sum_{i=1}^n\lambda_i(g(X)_i-c_i) + \sum_{k=1}^m\mu_k(h(X)_k)$$ where $h(x)_k$ is the k-th inequality condition.

and the minimization condition becomes $$\nabla L = 0$$

In addition to this regularization there are some important conditions called the Kahn-Karush-Tocker conditions(KKT conditions) which apply when dealing with inequality constraints.

just like in the case of the equality constrainsts we supply the inequality constraints as a dictionary.


A function which has a global minimum is called a convex function.

If it has a global maximum it is a concave function.

But no local minima or maxima should be there. basically rate of change of slope should always be increasing or decreasing.

# machine learning

* we specify a model that takes some input and based on the relation/patterns specified in that data tunes some parameters which are internal to the model based on which it can categorize or analyse some new data, and this information about the new data is the output.

* this using the input data to tune the model parameters is called "training the Model"

* there are three types of learning Algorithms:
1. Supervised(Labelled data)
the output that is to be generated given a certain data point, is passed along with the traing dataset. that is along with all the other features of a data point(for a student, features can be enrollemnt, viva marks, attendence, class-test marks, etc and if we are trying to predict the CGPA, the for the training datapoints, CGPA will be supplied as one of the features), the label will also be supplied as a feature.
2. Unsupervised(no grouping)
The labeles are no specified as part of the labelled data,  
3. Reinforcement
Tries all different paths which it can interacts with in a number of ways and based on the outcome of each of these interactions it may face a penalty based on one outcome and not on the other. Over many iterations it chosses the least penalized path. The design of what interactions are possible and which otcomes will be penalized is what is done from our side to get the desirable result.

$$Y^{\^}= \sum_{i}a_i X_i  + b$$

loss function, $$ \sum_{i} (Y_i - Y^{\^}_i)^2$$
$Y^{\^}$ is the predictid value, $Y$ is the actual value supplied as part of the training data, 

we want to minimize the loss function, wrt. the model parameter set(weights), $[a_1, a_2, a_3 \cdots a_i]$ and the bias $b$ and we can also have some constraints on the parameter set. 

Therefore it turns into am Optimization Problem.

we can have different types of linear models which impose different condition on the parameter set.

linear regression_ does not have any parameter(weight) constraint- simple least squares optimization problem.
